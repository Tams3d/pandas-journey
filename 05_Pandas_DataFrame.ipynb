# Pandas DataFrame


```python
import pandas as pd
```

## Creating DataFrame


### DataFrame from Python List


```python
student_data = [
    [100,80,10],
    [90,70,6],
    [80,60,5],
    [120,90,15]
]  # fmt:skip
pd.DataFrame(student_data, columns=["iq", "marks", "package"])
```

### DataFrame from Python Dictionary


```python
student_dict = {
    "iq": [100, 90, 80, 120, 100],
    "marks": [80, 70, 60, 90, 80],
    "package": [10, 6, 5, 15, 10],
}
students = pd.DataFrame(student_dict)
```
