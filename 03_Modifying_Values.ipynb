# Modifying Values


```python
import pandas as pd
```

```python
marks = pd.Series({"MAT": 97, "CHEM": 97, "PHY": 95, "CSC": 100})
marks
```

## Using Index


```python
marks[0] = 100
marks[1:3] = 97
marks
```

```python
marks[[1, 3]] = [95, 95]
marks
```

```python
marks["CSC"] = 97
marks
```

```python
marks["ENG"] = 93
marks
```

## Copy & View


```python
marks
```

```python
new_marks = marks.head(3)  # new_marks is a view of marks
new_marks
```

```python
marks["CSC"] = 100  # changes affects new_marks also
marks
```

```python
new_marks
```

```python
new_marks["PHY"] = 95  # changes affects marks also
new_marks
```

```python
marks
```

```python
new_marks = marks.head(3).copy()
new_marks
```

```python
marks["CHEM"] = 97
marks
```

```python
new_marks  # since new_marks is a copy, it is not affected
```
