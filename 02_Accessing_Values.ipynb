# Accessing Values


```python
import pandas as pd
```

```python
vk_df = pd.read_csv("data/kohli_ipl.csv", index_col="match_no")
vk = vk_df.iloc[:, 0]
vk
# .iloc[:, 0] is used to convert df to series (squeeze = True is deprecated)
```

```python
movies = pd.read_csv("data/bollywood.csv", index_col="movie")
movies = movies.iloc[:, 0]
```

## Indexing

- Does not support negative indexing


```python
x = pd.Series(
    [
        15,  # 0
        35,  # 1
        28,  # 2
        69,  # 3
        57,  # 4
        21,  # 5
    ]
)

print(x[4])
```

```python
# movies[index] will be deprecated (Future Warning)
movies[1]
```

### Fancy Indexing


```python
vk[[5, 10, 14, 27, 18, 1]]
```

### Indexing with Labels


```python
movies["Madras Cafe"]
```

### Boolean Indexing


```python
vk[vk >= 50].size
```

```python
movies.value_counts()[movies.value_counts() > 20]
```

## Slicing


```python
movies[5:16]
```

## Iterating


```python
# Iterates over values by default
for movie in movies.head():
    print(movie)
```

```python
# use series.index for iterating over indices
for movie in movies.head().index:
    print(movie)
```

```python
movies.value_counts().head(25).plot(kind="bar")
```
