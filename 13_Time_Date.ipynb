# Time and Date


```python
import datetime as dt

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
exps = pd.read_csv("data/expense_data.csv")
```

## Timestamp Object


```python
x = pd.Timestamp("2025-6-19 8:00PM")
x
```

### Attributes


```python
print("Year:", x.year)
print("Month Number:", x.month)
print("Day Number:", x.day)
print("Month:", x.month_name())
print("Day:", x.day_name())
print("Hour:", x.hour)
print("Minute:", x.minute)
print("Second:", x.second)
```

## Numpy DateTime


```python
date = np.datetime64("today")
date + np.arange(12)
```

## DatetimeIndex Object


```python
# From String
pd.DatetimeIndex(
    [
        "2023/1/1",
        "2022/1/1",
        "2021/1/1",
    ]
)
```

```python
# From datetime
pd.DatetimeIndex(
    [
        dt.datetime(2023, 1, 1),
        dt.datetime(2022, 1, 1),
        dt.datetime(2021, 1, 1),
    ]
)
```

```python
# From Timestamp
dt_index = pd.DatetimeIndex(
    [
        pd.Timestamp(2023, 1, 1),
        pd.Timestamp(2022, 1, 1),
        pd.Timestamp(2021, 1, 1),
    ]
)
dt_index
```

### Using as Index


```python
pd.Series([1, 2, 3], index=dt_index)
```

### Date Range


```python
pd.date_range(start="2025 6 18", end="2025 6 30")
```

```python
# Alternate days
pd.date_range(start="2025 6 16", end="2025 6 30", freq="2D")
```

```python
# Business days
pd.date_range(start="2025 6 16", end="2025 6 30", freq="B")
```

```python
# Particular day
pd.date_range(start="2025 6 16", end="2025 6 30", freq="W-TUE")
```

```python
# Hourly Data
pd.date_range(start="2025 6 16", end="2025 6 30", freq="h")

# Every 12 hours
pd.date_range(start="2025 6 16", end="2025 6 30", freq="12h")
```

```python
# Month End
pd.date_range(start="2025 6 1", end="2025 7 30", freq="MS")
```

```python
# Month End
pd.date_range(start="2025 6 1", end="2025 7 31", freq="ME")
```

```python
# Year Start
pd.date_range(start="2025 1 1", end="2030 1 30", freq="YS")
```

```python
# Year End
pd.date_range(start="2025 1 1", end="2030 1 30", freq="YE")
```

### Using Period


```python
# Number of days from start
pd.date_range(start="2025 6 18", periods=25)
```

## To Datetime


```python
s = pd.Series(["2023/1/1", "2022/1/1", "2021/100/1"])
pd.to_datetime(s, errors="coerce")
```

```python
pd.to_datetime(s, errors="coerce").dt.day_name()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(exps["Date"], exps["INR"])
```

```python
# Day wise bar chart
exps["Day_Name"] = exps["Date"].dt.day_name()
exps.groupby(["Day_Name"])["INR"].sum().plot(kind="bar")
```

```python
# Month wise bar chart
exps["Month_Name"] = exps["Date"].dt.month_name()
exps.groupby(["Month_Name"])["INR"].sum().plot(kind="bar")
```
