# Pandas Strings


```python
import pandas as pd
```

```python
ship = pd.read_csv("data/titanic.csv")
ship.head()
```

```python
ship["Name"].str.title()
```

```python
# Longest passenger name
ship["Name"][ship["Name"].str.len().argmax()]
```

```python
# Lastname of passenger
ship["Lastname"] = (
    ship["Name"].str.split(",").str.get(0).str.strip().str.title()
)
ship["Lastname"]
```

```python
ship[["Title", "Firstname"]] = (
    ship["Name"]
    .str.split(",")
    .str.get(1)
    .str.strip()
    .str.split(n=1, expand=True)
)
ship.head()
```

```python
ship["Title"].value_counts()
```

## Replace


```python
# replace
ship["Title"] = ship["Title"].replace(
    {"Ms.": "Miss", "Miss.": "Miss", "Mlle.": "Miss"}
)
ship["Title"].value_counts()
```

## Filtering


```python
# filtering
ship[ship["Firstname"].str.startswith("A")].head()
```

```python
# contains
ship["Firstname"][ship["Firstname"].str.contains("john", case=False)]
```

## Regex


```python
# regex - lastname start and end with vowels
ship["Lastname"][ship["Lastname"].str.contains("^[aeiouAEIOU].+[aeiouAEIOU]$")]
```
