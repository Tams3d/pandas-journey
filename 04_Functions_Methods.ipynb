# Functions & Methods


```python
import numpy as np
import pandas as pd
```

```python
vk_df = pd.read_csv("data/kohli_ipl.csv", index_col="match_no")
vk = vk_df.iloc[:, 0]
vk
```

```python
movies = pd.read_csv("data/bollywood.csv", index_col="movie").iloc[:, 0]
movies
```

```python
marks = pd.Series({"MAT": 97, "CHEM": 97, "PHY": 95, "CSC": 100})
marks
```

```python
subs = pd.read_csv("data/subs.csv").iloc[:, 0]
subs
```

## Data Methods


### Head


```python
print("First 5 Rows:\n", vk.head())  # returns first 5 rows by default
print("\nFirst 3 Rows:\n", vk.head(3))
```

### Tail


```python
print("Last 5 Rows:\n", vk.tail())  # returns last 5 rows by default
print("\nLast 3 Rows:\n", vk.tail(3))
```

### Random


```python
print("Random Movie:\n", movies.sample())  # returns 1 row randomly
print("\nRandom 5 Movies:\n", movies.sample(5))
```

### Value Counts

- Gives Frequency of data


```python
movies.value_counts()
```

### Sort Values


```python
vk.sort_values(ascending=False).to_numpy()
# ascending = True by default
# df.values is not recommended by pandas, to_numpy() returns values as numpy array
```

#### Inplace = True

- Makes changes to the original data
- Does not return anything


```python
vk.copy().sort_values(inplace=True)
vk

# copy() creates a copy of the df, used here to avoid error (view of some array)
```

### Sort Index


```python
movies.sort_index()
```

### Changing Data Type


```python
vk.astype("uint8")
```

### Between

- Both upper and lower limits are included


```python
vk[vk.between(75, 100)]
```

### Clip


```python
subs.clip(100, 200)
```

### Duplicates


```python
values = pd.Series([1, 3, 5, 5, 9, 4, 1, 7, 6, 3])
values
```

```python
# values with are duplicated
values[values.duplicated()]
```

#### Drop Duplicates

- Returns new Series
- Does not affects original


```python
values.drop_duplicates()
```

```python
values.drop_duplicates(keep="last")  # removes first duplicate occurrence, keeps last
```

### Missing Values


```python
temp = pd.Series([1, 4, None, 5, np.nan, 0])
temp
```

#### Is Null / Is NaN / Is NaT


```python
# series.isna() is preferred over series.isnull()
temp.isna()
```

#### Drop Missing Values


```python
temp.dropna()
```

#### Fill Missing Values


```python
# fillna() will get deprecated
temp.fillna(0)
```

```python
# fills NaN values from preceding value
temp.ffill()
```

```python
# fills NaN values from succeeding value
temp.bfill()
```

### Is In

- Checks in values


```python
vk[vk.isin([1, 2])]
```

### Apply


```python
movies.apply(str.capitalize)
```

```python
vk.apply(lambda runs: "Good Day" if runs > vk.mean() else "Bad Day")
```

## Series with Python Functionality


```python
values = pd.Series([10, 25, None, 37, 54, np.nan])

print("Total Count:", len(values))  # counts all values
print("Type:", type(values))
print("Sorted:", sorted(values))  # output in list
print("Minimum:", min(values))
print("Maximum:", max(values))

# print("\nAll attributes and methods:\n", dir(values))
```

### Type Conversion


```python
print("Data:", marks)
print("List:", list(marks))
print("Dictionary:", dict(marks))
```

### Membership Operators


```python
"Madras Cafe" in movies  # works with index for Series
```

```python
"Amitabh Bachchan" in movies.to_list()  # movies.values
```

## Math Methods


```python
# count - counts total number of values, does not include None, null, nan
print("Count:", vk.count())
print("Sum:", vk.sum())
print("Product:", vk.prod())
print("Minimum:", vk.min())
print("Maximum:", vk.max())
```

## Statistical Methods


```python
print("Average:", vk.mean())
print("Median:", vk.median())
print("Mode:", movies.mode())
print("Standard Deviation:", vk.std())
print("Variance:", vk.var())
```
