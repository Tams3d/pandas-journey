# Group By


```python
import pandas as pd
```

```python
movies = pd.read_csv("data/imdb-top-1000.csv")
ipl = pd.read_csv("data/deliveries.csv")
```

```python
genres = movies.groupby("Genre")
```

```python
# Applying aggregate functions on groupby
genres.sum()
```

```python
genres.mean(
    numeric_only=True
)  # numeric_only=True prevents Error caused by string values
```

```python
# Top 3 genres by total earning
genres["Gross"].sum().sort_values(ascending=False).head(3)
```

```python
# Genre with highest avg IMDB rating
genres["IMDB_Rating"].mean().sort_values(ascending=False).head(1)
```

```python
#  Director with most popularity
movies.groupby("Director")["No_of_Votes"].sum().sort_values(
    ascending=False
).head(1)
```

```python
movies.groupby("Star1")["Series_Title"].count().sort_values(ascending=False)
```

## Loop


```python
for group, data in genres:
    print(data)
```

```python
# Highest rated movie in each genre
for group, data in genres:
    print(data[data["IMDB_Rating"] == data["IMDB_Rating"].max()])
```

## Split Apply Combine


```python
genres.apply(min)
```

```python
# Number of movies starting with A from each genre
def foo(group):
    return group["Series_Title"].str.startswith("A").sum()
```

```python
genres.apply(foo)
```

```python
# Rank of each movie in group based on IMDB score
def rank_movie(group):
    group["genre_rank"] = group["IMDB_Rating"].rank(ascending=False)
    return group
```

```python
def norm(group):
    group["norm_rating"] = (
        group["IMDB_Rating"] - group["IMDB_Rating"].min()
    ) / (group["IMDB_Rating"].max() - group["IMDB_Rating"].min())
    return group
```

```python
genres.apply(rank_movie)
```

```python
genres.apply(norm)
```

## GroupBy Multiple


```python
duo = movies.groupby(["Director", "Star1"])
duo.get_group(("Aamir Khan", "Amole Gupte"))
```

```python
# Most earned actor director combo
duo["Gross"].sum().sort_values(ascending=False).head(1)
```

```python
# Best actor genre combo in terms of meta score
movies.groupby(["Star1", "Genre"])["Metascore"].mean().sort_values(
    ascending=False
).reset_index()
```

```python
ipl.head()
```

```python
# Top 10 batsman in terms of runs
ipl.groupby(["batsman"])["batsman_runs"].sum().sort_values(
    ascending=False
).head(10)
```

```python
# Batsman with maximum sixes
ipl[ipl["batsman_runs"] == 6].groupby("batsman")[
    "batsman_runs"
].count().sort_values(ascending=False)
```

```python
# Top 5 Batsman with most number of 4 and 6 in last 5 over
ipl[(ipl["over"] > 15) & (ipl["batsman_runs"].isin([4, 6]))].groupby(
    "batsman"
)["batsman_runs"].count().sort_values(ascending=False).head(5)
```

```python
# VK's records against all teams
ipl[ipl["batsman"] == "V Kohli"].groupby("bowling_team")["batsman_runs"].sum()
```

```python
# Function that can return highest score in any batsman
def highest(batsman):
    return (
        ipl[ipl["batsman"] == batsman]
        .groupby("match_id")["batsman_runs"]
        .sum()
        .sort_values(ascending=False)
        .head(1)
        .to_numpy()[0]
    )
```

```python
highest("V Kohli")
```
